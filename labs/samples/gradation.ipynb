```python
from math import pi
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import *

output_notebook()

p = Plot(width=400, height=400, x_range=Range1d(0, 1.2), y_range=Range1d(0, 1.2),
         toolbar_location=None)
p.add_glyph(ImageURL(url=['http://localhost:8000/pattern.png'], x=0, y=.05, w=1, h=.1, angle=pi/6))
show(p)
```

```python
p = figure(x_range=(0, 3.5), y_range=(0, 3.5))

hatch = dict(hatch_extra = {'gradation': ImageURLTexture(url='http://localhost:8000/pattern.png')},
             hatch_pattern=dict(value='gradation'),
             hatch_scale=256,
             hatch_color='navy')
p.quad(top=[2], bottom=[1], left=[1], right=[2],
       fill_color='white',
       **hatch)
p.quad(top=[3], bottom=[2], left=[2], right=[3],
       fill_color='white',
       **hatch)

show(p)
```
